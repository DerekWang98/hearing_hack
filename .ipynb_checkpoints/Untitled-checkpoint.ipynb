```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("uncleaned_data.csv")
```

```python
columns = df.columns
```

I'm going to be looking at the PTA results because they seem to be the most complete (around 100 rows have NaN values)

```python
# we look at the PTA tests
pta_columns = [c for c in columns if "PTA" in c ]
```

```python
no_nan = df[pta_columns].dropna(axis=0)
```

```python
gender = df.iloc[no_nan.index]['General_Gender']
adult_indigenous =  df.iloc[no_nan.index]['Adults_Questioanire_AboriginalTorresStraitIslander'].fillna(0)
child_indigenous =  df.iloc[no_nan.index]['SchooAge_Questioanire_AboriginalTorresStraitIslander'].fillna(0)
is_indigenous = adult_indigenous + child_indigenous
```

We are going to look specifically at the hearing ability of the indigenous population

```python
child_indigenous.value_counts()
```

```python
adult_indigenous.value_counts()
```

My guess is that class "2.0" represents the indigenous populationm counts 1 and 0 represent non (or non respondants)

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(no_nan)
```

```python
pca = PCA(n_components=2)
X_trans = pca.fit_transform(scaled)
```

The PTA test has 7 dimensions. I've used PCA to reduce $7\rightarrow2$ for easier visualisation

```python
plt.scatter(X_trans[:,0][is_indigenous!=2],
            X_trans[:,1][is_indigenous!=2],
            c="orange",label="Non indiogenous")
plt.scatter(X_trans[:,0][is_indigenous==2],
            X_trans[:,1][is_indigenous==2],
            c="blue",label="Indigenous")
plt.legend()
```

By the looks of it, it seems that the indigenous population have hearing ability that is most inline with the population

```python
from sklearn.covariance import EllipticEnvelope
envel = EllipticEnvelope()
envel.fit(X_trans)
envel.decision_function(X_trans)
```

```python
subfig1 = plt.subplot(1, 1, 1)
subfig1.scatter(X_trans[:,0][gender==0],
            X_trans[:,1][gender==0],
            c="orange",label="Female")
subfig1.scatter(X_trans[:,0][gender==1],
            X_trans[:,1][gender==1],
            c="green",label="Male")
# Show contours of the distance functions
xx, yy = np.meshgrid(np.linspace(plt.xlim()[0], plt.xlim()[1], 100),
                     np.linspace(plt.ylim()[0], plt.ylim()[1], 100))
zz = np.c_[xx.ravel(), yy.ravel()]

mahal_emp_cov = envel.mahalanobis(zz)
mahal_emp_cov = mahal_emp_cov.reshape(xx.shape)
emp_cov_contour = subfig1.contour(xx, yy, np.sqrt(mahal_emp_cov),
                                  cmap=plt.cm.PuBu_r,
                                  linestyles='dashed')
plt.legend()
```

Looking at the plot above, you can see that there's a region of "normal" hearing ability (where the points are most dense on the left hand side) and then it spreads towards the right hand side. 

So perhaps we have a measure of what is considered as "abnormal" hearing: data points that land outside the innermost ellipse are those
